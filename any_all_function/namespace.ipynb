## Namespace
### A namespace is place where python store variable and functions name with their values.
### example:- x = 10


## Types of namespace
### 1. local namespace
### 2. global namespace
### 3. builtin namespace

```python
# local namespace
# names created insida a functions.
def test():
    a = 10
    print(locals())
test()
```

```python
# global namespace
# names created outside of all functions.
a = 10
def test():
    x = 20
    print(x)
print(globals())
```

```python
# built in namespace
# names given by python itself.
import builtin
print(dir(builtin))
```
