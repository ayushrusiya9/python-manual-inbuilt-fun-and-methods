## any() 
### any(itrable) return True if at least one element in itrable is true. its like 'or' operator.

```python
# any example
any(['python',1,0])
```

# all()
### all(itrable) return True only if all elements in itrable is true. its like 'and' operator.

```python
# example all()
all([1,'python',1])
```
