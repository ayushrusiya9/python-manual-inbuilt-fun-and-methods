```python
# instance varibale example
# self dependend veriable ko bhi inststance veriable bhi kehte hai 
class Student:
    def __init__(self, name, roll):
        self.n = name        
        self.r = roll        

s1 = Student("Ayush", 101)
s2 = Student("Rahul", 102)

print(s1.n, s1.r)   
print(s2.n, s2.r)  
```

```python
# class veriable / static
class Student:
    school_name = "Cybrome"  

    def __init__(self, name):
        self.name = name         

s1 = Student("Ayush")
s2 = Student("Rahul")

print(s1.school_name)   
print(s2.school_name)   
print(Student.school_name) 
```
