## Scope
### scope is decide where the variable can be used (accessed) in the code.
## Types of scope
### 1. local scope
### 2. global scope
### 3. nonlocal scope

```python
# local scope
# variable usable only inside a function.
def fun():
    a = 10
    print(a)
fun()
print(a) # this give error -> name a is not defined
```

```python
# global scope
# veriable usable everywhere in file.
x = 10

def fun():
    print(x)

fun()
```

```python
# non local scope
# used in nasted function to change outer function variable.
def outer():
    x = 10
    def inner():
        nonlocal x
        x = 20

    inner()
    print(x)
outer()
```
