# Higher order function
### a function that accepts or return another function.

```python
def apply(func, x):
    return func(x)

def square(x):
    return x*x

print(apply(square, 2))
```

## Types of higher order function 
#### 1. map() - map(func, itrable,...)
#### 2. filter() - filter(func, itrable)
#### 3. reduce() - reduce(func, itrable) this is in functool module
#### 4. decorator() - a function that modify behaviour of function without changing there code.
